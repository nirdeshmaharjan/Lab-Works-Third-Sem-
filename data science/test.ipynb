```python
def prime(n):
    c=0
    for i in range(2,n+1):
        if(n%i==0):
            c=c+1
    if(c==1):
        return True
    else:
        return False
n=int(input("enter the number :"))
if(prime(n)):
    print("prime")
else:
    print("not prime")
```

# visualizing data

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

# line chart

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 54.3, 99075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='blue', marker='1', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
```

```python
flight = sns.load_dataset('flights') # load flight dataset
print(flight)

flight_new = flight.groupby('year').sum(numeric_only=True).reset_index()
# line chart
sns.lineplot(x='passengers',y='year',data=flight_new)
```

# bar graph

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [15, 5, 5, 5, 15]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)

plt.title("My Favorite Movies")     # add a title
plt.ylabel("# of Academy Awards")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./titanic.csv")
df.head(8
```

```python
# countplot (Survived data) Dataset Titanic
sns.countplot(data=df, x='Sex')
df['Survived'].value_counts()
```

```python
# Bivariate Analysis
sns.barplot(x='Pclass',y='Age',hue='Sex',data=df)
```

```python
 # pie chart

df['Sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python

# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['Age'])
```

```python

# Bivarient Analysis


#Box plot (Numerical - Categorical)

sns.boxplot(x='Sex',y='Age',hue='Survived',data=df)
```

```python
# Bivariant analysis
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Died')
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Survived')

plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
